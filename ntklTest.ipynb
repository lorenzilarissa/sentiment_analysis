```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Substitua o URL abaixo pelo URL do arquivo CSV no seu reposit√≥rio do GitHub
github_csv_url = 'C:\\Users\\larissa.lorenzi\\Git\\ai-notebooks\\SentimentAnalysisBlubot/chat_data.csv'

# Use o pandas para ler o arquivo CSV a partir do URL
df = pd.read_csv(github_csv_url)
```

```python
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
```

```python
# Vamos criar uma nova coluna chamada "sentiment_score" para armazenar os valores de sentimento
df['sentiment_score'] = df['message'].apply(lambda message: sid.polarity_scores(message)['compound'])
```

```python
df = df.sort_values(by='sentiment_score', ascending=False)
df.head(20)
```

```python
def analyse_sentment(message):
    sentment = sid.polarity_scores(message)
    if sentment['compound'] >= 0.05:
        return 'Positivo'
    elif sentment['compound'] <= -0.05:
        return 'Negativo'
    else:
        return 'Neutro'
```

```python
df['sentment'] = df['message'].apply(analyse_sentment)
```
