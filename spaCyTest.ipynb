```python
sia = SentimentIntensityAnalyzer()

positive_count = 0
negative_count = 0

for sentence in sentences:
    sentiment_scores = sia.polarity_scores(sentence)
    sentiment = sentiment_scores['compound']

    if sentiment >= 0.05:
        positive_count += 1
        print('Mensagem positiva: ', sentence)
    elif sentiment <= -0.05:
        negative_count += 1
        print('Mensagem negativa: ', sentence)
    else:
        print('Mensagem neutra: ', sentence)
```
